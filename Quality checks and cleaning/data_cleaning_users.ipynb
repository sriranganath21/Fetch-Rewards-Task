```python
import pandas as pd
import json
from pprint import pprint
```

```python
#reading the json file
with open(r'C:\Users\srira\OneDrive\Desktop\fetchrewards\users.json') as file:
    users = json.loads("[" + 
        file.read().replace("}\n{", "},\n{") + 
    "]")

    pprint(users)
```

```python
# normalizing the json 
users_df = pd.json_normalize(users)
users_df
```

```python
#renaming the flattned columns names
users_clean = users_df.rename(columns={"_id.$oid": "user_id", "createdDate.$date": "created_date","lastLogin.$date":"lasLogin_date" })
users_clean
```

```python
#Most of the data is from Consumers
users_clean.groupby("role").agg({"user_id": pd.Series.nunique})
```

```python
#Most of the signups is through emails
users_clean.groupby("signUpSource").agg({"user_id": pd.Series.nunique})
```

```python
#Removing the data related to role Fetch Staff
users_clean = users_clean[users_clean.role != "fetch-staff"]
users_clean
```

```python
#Removing the null values, we lost around 70 rows of data
users_clean = users_clean.dropna()
users_clean
```

```python
#There are somany rows of duplicate dat for a given user. 
users_clean[users_clean.user_id == "5ff1e194b6a9d73a3a9f1052"]
```

```python
#Filtering the dataset based on User ID
users_clean = users_clean.drop_duplicates(subset='user_id').reset_index(drop=True)
users_clean
```

```python
#Reindexing
users_clean = users_clean.reindex(columns=['user_id','created_date','state','active','signUpSource','lasLogin_date'])
users_clean
```

```python
#Converting the date fields to Datetime formats
users_clean['created_date'] = pd.to_datetime(users_clean['created_date'], unit="ms")
users_clean['lasLogin_date'] = pd.to_datetime(users_clean['lasLogin_date'], unit="ms")
users_clean
```

```python
#exporting it to csv
users_clean.to_csv('users_clean.csv',index = False)
```
