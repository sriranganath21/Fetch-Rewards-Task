```python
import pandas as pd
import json
from pprint import pprint
```

```python
#reading the json file
with open(r'C:\Users\srira\OneDrive\Desktop\fetchrewards\brands.json') as file:
    brands = json.loads("[" + 
        file.read().replace("}\n{", "},\n{") + 
    "]")

    pprint(brands)
```

```python
# normalizing the json 
brands_df = pd.json_normalize(brands)
brands_df
```

```python
#renaming the flattned columns names
brands_clean = brands_df.rename(columns={"_id.$oid": "brand_id","cpg.$ref":"cpg_ref" })
brands_clean
```

```python
brands_clean = brands_clean.drop_duplicates(subset='brand_id').reset_index(drop=True)
brands_clean
```

```python
#Exporting it to
brands_clean.to_csv('brands_clean.csv',index = False)
```
